```r
# install.packages('readxl')
library(readxl)
data <- read_excel('data tugas mtp.xlsx')
```

```r
data_update <- data[c(3, 4, 5, 8, 9, 10, 11, 14)] #AMBIL VARIABEL AHH, IPM, PERSENTASE PENDUDUK MISKIN, TPT, JUMLAH PENDUDUK, PERTUMBUHAN EKONOMI, GINI RATIO, RATA-RATA LAMA SEKOLAH
```

```r
datatable(data_update, caption = "Variabel")
```

```r
data_standardized <- round(scale(data_update),4)
```

```r
data_standardized
```

```r
datatable(data_standardized, caption = "Data Hasil Standardisasi")
```

```r
data_standardized
```

```r
jumlah_klaster <- c(1:12) #Vektor yang berisikan jumlah klaster yang ingin dilihat nilai dari total within-cluster sum of squares
```

```r
jumlah_klaster
```

```r
within_ss <- c() #Vektor kosong yang akan diisi nilai total withincluster sum of squares
```

```r
for (i in jumlah_klaster) {
 within_ss <- c(within_ss, kmeans(x = data_standardized, centers = 
i, nstart = 25)$tot.withinss)
}

plot(x = jumlah_klaster, y = within_ss, type = "b", xlab = "Number of 
Cluster",
 ylab = "Total Within Sum of Squares", main = "Elbow Plot")
abline(v = 8, col = 'red')
```

```r
set.seed(123)
kmeans_clustering <- kmeans(x = data_standardized, centers = 8, nstart = 25) #parameter nstart digunakan untuk memberitahu fungsi  berapa kali inisiasi centroid awal (secara acak) yang akan dibentuk.
```

```r
kmeans_clustering
```

```r
data_update %>% mutate(Klaster = kmeans_clustering$cluster) %>% group_by(Klaster) %>% summarise(Mean_AHH = mean(AHH), Mean_IPM = mean(IPM), Mean_PPM = mean(Persentase.Penduduk.Miskin), Mean_TPT = mean(TPT), Mean_JP = mean(Jumlah.Penduduk), Mean_PE = mean(pertumbuhan.ekonomi), Mean_gini_ratio = mean(gini.ratio), Mean_rata_rata_lama_sekolah = mean(Rata.rata.Lama.Sekolah))
```

```r
data %>%
 mutate(Klaster = kmeans_clustering$cluster) %>%
 select('Kab.Kota', Klaster) %>%
 arrange(Klaster)
View(data %>%
 mutate (Klaster = kmeans_clustering$cluster) %>%
 select('Kab.Kota', Klaster) %>%
 arrange(Klaster))
```
